```python
import csv
import random
from datetime import datetime, timedelta


departments = [
    ("D101", "Information Technology", "New York"),
    ("D102", "Human Resources", "San Francisco"),
    ("D103", "Research and Development", "Seattle"),
    ("D104", "Marketing", "Chicago"),
    ("D105", "Sales", "Austin"),
    ("D106", "Finance", "Boston"),
    ("D107", "Customer Service", "Denver"),
    ("D108", "Logistics", "Atlanta"),
    ("D109", "Legal Department", "Washington D.C."),
    ("D110", "Public Relations", "Los Angeles")
]

# Helper functions
def random_date(start, end):
    return (start + timedelta(days=random.randint(0, (end - start).days))).strftime('%Y-%m-%d')

def random_phone_number():
    return f"{random.randint(100,999)}-{random.randint(100,999)}-{random.randint(1000,9999)}"

employees = []
for i in range(1, 501):
    employee_id = "E" + str(1000 + i)

    first_name = random.choice([
    "James", "Mary", "John", "Patricia", "Robert", "Jennifer", "Michael", "Linda", 
    "William", "Elizabeth", "David", "Barbara", "Richard", "Susan", "Joseph", "Jessica", 
    "Thomas", "Sarah", "Charles", "Karen", "Christopher", "Nancy", "Daniel", "Lisa", 
    "Matthew", "Margaret", "Anthony", "Betty", "Mark", "Sandra", "Steven", "Ashley", 
    "Paul", "Dorothy", "Andrew", "Kimberly", "Joshua", "Emily", "Kevin", "Donna", 
    "Brian", "Michelle", "George", "Carol", "Edward", "Amanda", "Ronald", "Melissa", 
    "Timothy", "Deborah", "Jason", "Stephanie", "Jeffrey", "Rebecca", "Ryan", "Laura", 
    "Jacob", "Helen", "Gary", "Sharon", "Nicholas", "Cynthia", "Eric", "Kathleen", 
    "Jonathan", "Amy", "Stephen", "Shirley", "Larry", "Angela", "Justin", "Anna", 
    "Scott", "Ruth", "Brandon", "Brenda", "Benjamin", "Pamela", "Samuel", "Nicole", 
    "Gregory", "Katherine", "Frank", "Samantha", "Alexander", "Christine"
])
    last_name = random.choice([
    "Smith", "Johnson", "Williams", "Brown", "Jones", "Garcia", "Miller", "Davis", 
    "Rodriguez", "Martinez", "Hernandez", "Lopez", "Gonzalez", "Wilson", "Anderson", 
    "Thomas", "Taylor", "Moore", "Jackson", "Martin", "Lee", "Perez", "Thompson", 
    "White", "Harris", "Sanchez", "Clark", "Ramirez", "Lewis", "Robinson", "Walker", 
    "Young", "Allen", "King", "Wright", "Scott", "Torres", "Nguyen", "Hill", 
    "Flores", "Green", "Adams", "Nelson", "Baker", "Hall", "Rivera", "Campbell", 
    "Mitchell", "Carter", "Roberts", "Gomez", "Phillips", "Evans", "Turner", "Diaz", 
    "Parker", "Cruz", "Edwards", "Collins", "Reyes", "Stewart", "Morris", "Morales", 
    "Murphy", "Cook", "Rogers", "Gutierrez", "Ortiz", "Morgan", "Cooper", "Peterson", 
    "Bailey", "Reed", "Kelly", "Howard", "Ramos", "Kim", "Cox", "Ward", "Richardson", 
    "Watson", "Brooks", "Chavez", "Wood", "James", "Bennet", "Gray", "Mendoza", 
    "Ruiz", "Hughes", "Price", "Alvarez", "Castillo", "Sanders", "Patel", "Myers", 
    "Long", "Ross", "Foster", "Jimenez"
])
    department_id = random.choice(departments)[0]
    position = random.choice(["Analyst", "Manager", "Developer", "Designer", "Consultant", "Sales Representative", "Engineer"])
    email = first_name.lower() + "." + last_name.lower() + "@example.com"
    phone_number = random_phone_number()
    salary = random.randint(50000, 120000)
    hire_date = random_date(datetime(2010, 1, 1), datetime(2023, 1, 1))
    start_date = datetime.strptime(hire_date, '%Y-%m-%d')
    total_years_as_employee = (datetime.now() - start_date).days // 365
    leave_remaining = min(30, max(0, 30 - total_years_as_employee))
    age = random.randint(24, 60)
    gender = random.choice(["M", "F", "O"])
    previous_position = random.choice(["Junior " + position, "Senior " + position, "Intern"])
    highest_degree_achieved = random.choice(["Bachelors", "Masters", "PhD", "Diploma"])
    employment_status = random.choice(["Active", "On Leave", "Terminated", "Retired"])
    
    employees.append([
        employee_id, first_name, last_name, department_id, position, email, phone_number, salary,
        hire_date, total_years_as_employee, leave_remaining, age, gender, previous_position,
        highest_degree_achieved, employment_status
    ])

# Write to CSV files
with open('employees.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["employee_id", 
                     "first_name", 
                     "last_name", 
                     "department_id", 
                     "position", 
                     "email", 
                     "phone_number", 
                     "salary", 
                     "hire_date", 
                     "total_years_as_employee", 
                     "leave_remaining", 
                     "age", 
                     "gender", 
                     "previous_position", 
                     "highest_degree_achieved", 
                     "manager_id", 
                     "employment_status"])
    writer.writerows(employees)

with open('departments.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["department_id", 
                     "department_name", 
                     "location"])
    writer.writerows(departments)

print("CSV files have been created.")
```
